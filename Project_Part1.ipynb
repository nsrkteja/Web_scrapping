## Problem Statement: 
To Analyse how the Primary Price is related to other features

```python
# Loading required libraries

import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
# 1. stock type <p> --> class = 'stock-type'
# 2. car name <h2> --> class = 'title'
# 3. Primary price <span> --> class = 'primary-price'
# 4. EMI <span> --> class="js-estimated-monthly-payment-formatted-value-with-abr"
# 5. virtual Appoiments/ Home Delivery <span> --> class='sds-badge__label'
# 6. Dealer <div> --> class='dealer-name'
# 7. Rating <span> --> class="sds-rating__count"
# 8. Reviews <span> --> class="sds-rating__link sds-button-link"
# 9. Secondary price Drop <span> --> class="secondary-price price-drop"
# 10. Vehicle Badging <div> --> class="vehicle-badging"
```

```python
%%time

stockType = []
title = []
primaryPrice = []
priceDrop = []
emi = []
dealer = []
rating = []
reviewsNumber = []
badge = []

for i in range(1, 41):
    URL = 'https://www.cars.com/shopping/results/?page={}&page_size=20&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=mitsubishi&maximum_distance=20&mileage_max=&models[]=mitsubishi-outlander&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip='. format(i)
    
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div', attrs={'class' : 'vehicle-card'}):

        stock = x.find('p', attrs={'class' : 'stock-type'})
        if stock is None:
            stockType.append(np.NaN)
        else:
            stockType.append(stock.text)

        carName = x.find('h2', attrs={'class' : 'title'})
        if carName is None:
            title.append(np.NaN)
        else:
            title.append(carName.text)

        pp = x.find('span', attrs={'class' : 'primary-price'})
        if pp is None:
            primaryPrice.append(np.NaN)
        else:
            primaryPrice.append(pp.text)
        
        drop = x.find('span', attrs={'class':'price-drop'})
        if drop is None:
            priceDrop.append(np.NaN)
        else:
            priceDrop.append(drop.text)
        
        mp = x.find('span', attrs={'class':'js-estimated-monthly-payment-formatted-value-with-abr'})
        if mp is None:
            emi.append(np.NaN)
        else:
            emi.append(mp.text)
        
        dealerName = x.find('div', attrs={'class':'dealer-name'})
        if dealerName is None:
            dealer.append(np.NaN)
        else:
            dealer.append(dealerName.text)
        
        ratings = x.find('span',attrs={'class':'sds-rating__count'})
        if ratings is None:
            rating.append(np.NaN)
        else:
            rating.append(ratings.text)
        
        review = x.find('span',attrs={'class':'sds-rating__link sds-button-link'})
        if review is None:
            reviewsNumber.append(np.NaN)
        else:
            reviewsNumber.append(review.text)
        
        badging = x.find('div', attrs={'class':'vehicle-badging'})
        if badging is None:
            badge.append(np.NaN)
        else:
            badge.append(badging.text)
```

```python
print(len(stockType))
print(len(title))
print(len(primaryPrice))
print(len(priceDrop))
print(len(emi))
print(len(dealer))
print(len(rating))
print(len(reviewsNumber))
print(len(badge))
```

## Create a dataFrame

```python

df = pd.DataFrame({'New/Used' : stockType,
                   'Model' : title, 
                   'Primary Price' : primaryPrice,
                   'Secondary Price Drop': priceDrop,
                   'Vehicle Badge': badge,
                   'EMI': emi,
                   'Dealer Name': dealer,
                   'Rating': rating,
                   'Reviews': reviewsNumber,
                  })
```
